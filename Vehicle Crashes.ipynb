# Data Cleaning

The first step in cleaning the data will be to simply make the column names lowercase. This will improve readability and also make writing this code a bit easier.

```python
crashes.columns = map(str.lower, crashes.columns)
crashes = crashes.rename(columns={'crn':'crash_number', 'crash_year': 'year', 'crash_month': 'month'})
```

Next, we will fill any blank values with NaN and create a heatmap to determine what areas of the dataset have missing values.

```python
crashes = crashes.replace(r'^\s*$', np.nan, regex=True)
```

```python
#Import seaborn to utilize the heatmap function
import seaborn as sns
import matplotlib.pyplot as plt

#Create function to print a heatmap of null values in the dataset
def null_plot(df, figsize=(18,15)):
    plt.figure(figsize=figsize)
    df_null = df.isnull()
    sns.heatmap(~df_null, cbar=False, yticklabels=False)
    plt.xticks(rotation=90, size='x-large')
    plt.show()
    
null_plot(crashes)
```

We can see that there are many columns that are almost entirely null. Many of these are related to crashes that occur in work/construction zones. Because this is such a large amount of the data set and we are not specifically concerned with work zone related crashes, we will drop these columns from the dataset.

```python
#Remove columns related to work zone crashes
drop_columns = ['work_zone_type', 'work_zone_loc', 'cons_zone_spd_lim', 'workers_pres',
                'wz_close_detour', 'wz_flagger', 'wz_law_offcr_ind', 'wz_ln_closure',
                'wz_moving', 'wz_other', 'wz_shlder_mdn', 'rdwy_surf_type_cd', 'spec_juris_cd']

crashes = crashes.drop(columns=drop_columns)
```

# Heatmap of Crash Locations

Utilizing the Basemap package, we can create a heatmap to help visualize where crashes occur the most across the state.

```python
#Import Basemap package 
from mpl_toolkits.basemap import Basemap

#Create a new dataset to remove entries with missing coordinates
coordinates = crashes[(crashes['dec_long'].notnull()) & (crashes['dec_lat'].notnull())][['dec_lat', 'dec_long']]
```

```python
def pa_crash_map(longs, lats, cmap='hot_r', title='Pennsylvania Motor Vehicle Crash Frequency, 2018'):
    #Create a figure to display the entire state of Pennsylvania
    fig = plt.figure(num=None, figsize=(12,10))
    m = Basemap(llcrnrlat=39.654, urcrnrlat=42.342, llcrnrlon=-80.650, urcrnrlon=-74.649, resolution='h')

    #Define coordinates from the dataset
    x_plot = np.array(longs)
    y_plot = np.array(lats)

    #Create a hexbin overlay of the coordinates to show areas of higher frequency
    m.hexbin(x_plot, y_plot, mincnt=1, gridsize=75, cmap='hot_r')

    m.colorbar(location='bottom', label='Number of Crashes')

    #Create a dictionary of large Pennsylvania cities and their coordinates
    cities = {'Pittsburgh':(40.444, -79.997), 'Philadelphia':(39.946, -75.152),
              'Allentown':(40.600, -75.471), 'Reading':(40.334, -75.927),
              'Harrisburg':(40.253, -76.865), 'Erie':(42.128, -80.086), 
              'York':(39.963, -76.730), 'Lancaster':(40.039, -76.312), 
              'Scranton':(41.407, -75.663), 'Wilkes-Barre':(41.245, -75.872)}

    #Draw borders for counties, states, and coasts
    m.drawstates(linewidth=0.5, linestyle='solid', color='black')
    m.drawcounties(linewidth=0.5, linestyle='solid', color='grey')
    m.drawcoastlines(linewidth=0.5, linestyle='solid', color='black')
          
    #Plot and label the cities in the map
    for key in cities:
#         m.plot(cities[key][1], cities[key][0], latlon=True, marker='.', color='b')
        plt.text(cities[key][1]+.03, cities[key][0]+.03, '{}'.format(key), fontweight='bold', color='k')
    
    plt.title(title)
    
pa_crash_map(coordinates['dec_long'], coordinates['dec_lat'])

plt.show()
```

```python
crash_by_month = crashes['month'].value_counts().sort_index()
crash_by_month = pd.DataFrame(crash_by_month)
crash_by_month.reset_index(level=0, inplace=True)
crash_by_month.rename(columns={'index':'month', 'month':'count'}, inplace=True)
```

```python
#Import datetime package to convert numerical dates 
import datetime as dt

#Create a function to convert from integers to month names
def month_name(month):
    datetime_object = dt.datetime.strptime(str(month), '%m')
    month_name = datetime_object.strftime('%B')
    return month_name

#Apply the previously created formula to the month column to create 'month_name'
crash_by_month['month_name'] = crash_by_month['month'].apply(month_name)

crash_by_month['percentage'] = (crash_by_month['count'] / crash_by_month['count'].sum())
```

```python
#Create a bar chart contianing frequencies
fig1 = plt.figure(num=1, figsize=(5,5))
crash_by_month.plot(kind="bar", x='month_name', y='count', legend=False)
plt.axhline(crash_by_month['count'].mean(), color='orange')
plt.xlabel('Month')
plt.ylabel('Crashes')
plt.title('Pennsylvania Motor Vehicle Crashes by Month, 2018')

print(crash_by_month[crash_by_month['count'] > crash_by_month['count'].mean()]['percentage'].sum())
```

This bar chart shows that, as one would expect, vehicles crashes occur at an above average rate during the winter months (October - February), with 45% of the total crashes for the year occurring in these five months.
